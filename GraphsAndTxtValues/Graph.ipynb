```python
import matplotlib.pyplot as plt 
```

```python
def getInfoFromFiles(fileName, x_list, y_list):
    roster = list()
    with open(fileName, 'r') as file:
        lines = [line.rstrip('\n') for line in file]
        for line in lines:
            if(line.isdigit()):
                roster.append(int(line))
        x_list += roster[0::];
        
        roster = lines[1::2]
        for element in roster:
            y_list.append(float(element))
                        
                

```

```python
x_first_version = list()
y_first_version = list()

x_second_version = list()
y_second_version = list()

file1 = 'RegularInfo.txt'
file2 = 'OptimizedInfo.txt'

```

```python
getInfoFromFiles(file1, x_first_version, y_first_version)
```

```python
getInfoFromFiles(file2, x_second_version, y_second_version)
```

```python
plt.plot(x_first_version, y_first_version)
plt.plot(x_second_version, y_second_version)
```
